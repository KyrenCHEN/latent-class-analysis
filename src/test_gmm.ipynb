```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from gmm import GMM

from sklearn.datasets import make_blobs
```

## Generate synthetic data

```python
data = make_blobs(n_samples=1000, n_features=2, centers=3, cluster_std=[2,0.5,1], random_state=3)
x1 = data[0][:,0]
x2 = data[0][:,1]
labels = data[1]

fig,ax = plt.subplots(figsize=(12,7))
colors = ['#42A8DF','#E1A429','#E14343']
for i,l in enumerate(np.unique(labels)):
    ax = plt.scatter(x1[labels==l], x2[labels==l], s=80, c=colors[i])
plt.tight_layout()
```

```python
gmm = GMM(n_components=3)
```

```python
gmm.fit(data[0])
```
